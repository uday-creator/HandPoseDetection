```python

cap = cv2.VideoCapture(0)

#Now instantiating our media pipe hand model using with command
with mp_hands.Hands(min_detection_confidence = 0.8, min_tracking_confidence = 0.5,max_num_hands = 10) as hands:  #Taking 2 keyword arguments 1st is used for detectong hands and 2nd is used for tracking hands
    while cap.isOpened():
        ret, frame = cap.read() #storing what our camera is capturing



        #Coverting oolor
        image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) #Shifting color of frame

        #Set flag
        image.flags.writeable = False #Making our wrtable false sp that we can't copy oir image and our performance basically gets tunes

        # Detection
        res = hands.process(image)

        #set flag
        image.flags.writeable = True 

        #convrt color
        image = cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        print(res)

        #Rendering results
        if res.multi_hand_landmarks:
            for num, hand in enumerate(res.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS,
                                         mp_drawing.DrawingSpec(color = (121,22,76),thickness = 2,circle_radius = 6), #for landmarks
                                         mp_drawing.DrawingSpec(color = (250,44,250),thickness = 2,circle_radius = 2),#for line
                                         ) #for drawing the connections between the coordinates in the hands


        #Drwaing spec is a media pipe class that allows to change the color of your detections in this case we use it to cutmize the landmarks and connections.
        cv2.imshow("Hand Tracking", image) #showing our rendered image

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
 #Now inserting label on hands
def get_label(index,hand,res):
    output = None
    for idx,classification in enumerate(res.multi_handedness):
        if classification.classification[0].index == index:
            label = classification.classification[0].label
            score = classification.classification[0].score
            text = "{} {}".format(label,round(score, 2))

            #Extract coordinates
            coords = tuple(np.multiply(
                np.array( (hand.landmark[mp_hands.HandLandmark.WRIST].x,hand.landmark[mp_hands.HandLandmark.WRIST].y)),
            [640,480]).astype(int))

            output = text,coords
    return output
```

```python
#now calculate angles 

joint_list = [[8,7,6],[12,11,10],[4,3,2],[1,0,5]]  ##detecting angles 7 and 11
```

```python
def draw_finger_angles(image, res, joint_list):
    
    #loop through habds
    for hand in res.multi_hand_landmarks:
        #loop through joint sets
        for joint in joint_list:
            a = np.array([hand.landmark[joint[0]].x,hand.landmark[joint[0]].y]) #coordinates for joints in our hands (1st coord)
            b = np.array([hand.landmark[joint[1]].x,hand.landmark[joint[1]].y]) #vertex 2nd coord
            c = np.array([hand.landmark[joint[2]].x,hand.landmark[joint[2]].y])
            
            
            #Now converting our coord into radians and then converting into angles
            
            radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
            angle   = np.abs(radians*180.0/np.pi)
            
            if angle > 180.0:
                angle = 360 - angle
            
            cv2.putText(image,str(round(angle,2)), tuple(np.multiply(b, [600,400]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2, cv2.LINE_AA)
            
    return image
```
